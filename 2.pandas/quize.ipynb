```python
# 3-4 월별 평균을 출력해 보세요
df = pd.read_csv('vaccine.csv')
df["month"] = pd.DatetimeIndex(df["date"]).month
result = df[["month", 'subject', 'day1num','day1sum', 'day2sum', 'day1rate','day2num', 'day2rate']].groupby("month").mean()
result
```
