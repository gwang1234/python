```python
data = pd.read_excel('fine_dust.xlsx', index_col='area')
data2018 = data[2018]
data2018
```

```python
plt.figure(figsize=(15,4))
plt.plot(data2018, color='b', marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 Fine Dust Line Graph')
plt.show()
```

```python
plt.figure(figsize=(15,4))
for year in range(2016,2020):
    chartData = data[year]
    plt.plot(chartData, marker='s', label=year)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016~2019 Fine Dust Line Graph')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.ylim(35,45)
plt.bar(data2018.index, data2018, color='g')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 Fine Dust Line Graph')
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange(4)

plt.figure(figsize=(15,4))

data4 = data.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016,2020):
    chartData = data4[year]
    plt.bar(index, chartData, width=0.2, label=year)
    index = index+0.2
plt.ylim(35,55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016~2019 Fine Dust Line Graph')
plt.xticks(index-0.5, ["Gyeonggi", 'Incheon', 'Busan','Daegu'])
plt.grid()
plt.legend()
plt.show()
```
