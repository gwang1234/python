```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 2-1
result = tips.sort_values(['total_bill', 'tip'])
result.head(10)
```

```python
# 2-2
result = tips.sort_values(['size','tip'], ascending=[False, True])
result.head(10)
```

```python
titatic = pd.read_csv('train.csv')
titatic.head()
```

```python
# 4-1
titatic[['Sex', "Survived"]].groupby(['Sex']).mean().sort_values(by='Survived', ascending=False)
```

```python
# 4-2
titatic[['Parch', 'Survived']].groupby(['Parch'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
titatic.groupby(['Pclass', 'Sex'])[['Survived', 'Age']].mean()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=titatic)
```
